Vamos a analizar las licitaciones de Paraguay, del 2013 a julio 2026. Para eso usaremos el primer dataset que contiene los datos de los ítem adjudicados y el segundo que contiene la lista de las empresas adjudicadas.

```python
import pandas as pd
#Aquí cargo el archivo que contiene todas las adjudicaciones.
#Hay una fila por cada ítem adjudicado, no por cada licitación.
#Eso nos permite después matchear con todas las empresas que fueron adjudicadas.
df = pd.read_csv('awards.csv')
df.head()
```

```python
#Aquí subimos el segundo dataset que se trata de las empresas adjudicadas
df_proveedores = pd.read_csv('awa_suppliers.csv')
df_proveedores.head()
```

```python
#Ahora vamos a juntarlas
df_completo = pd.merge(
    df,                     
    df_proveedores,         
    on=['compiledRelease/id', 'compiledRelease/awards/0/id'],  
    how='left'       
)
```

```python
df_completo.sample(5)
```

```python
#Aquí vemos el resultado de las filas
filas_df = len(df)
filas_proveedores = len(df_proveedores)
filas_coincidencias = len(df_completo)

print(f"Filas en df: {filas_df}")
print(f"Filas en df_proveedores: {filas_proveedores}")
print(f"Filas en df_coincidencias: {filas_coincidencias}")
```

```python
#Ahora creamos una columna de año
#Primero pasamos la columna date a datetime y luego extraemos el año
fecha_convertida = pd.to_datetime(df_completo['compiledRelease/awards/0/date'], errors='coerce')
df_completo['year'] = fecha_convertida.dt.year

#Limpiamos el número para que no tenga comas y vemos el resultado
df_completo['year'] = df_completo['year'].fillna(0).astype('int64')
print(df_completo[['compiledRelease/awards/0/date', 'year']].head())
```

```python
df_completo.columns
```

```python
#Simplificamos las columnas
columnas_nuevas = {
    'compiledRelease/id': 'id_licitacion_global',
    'compiledRelease/awards/0/id': 'id_adjudicacion_contrato',
    'compiledRelease/awards/0/status': 'estado_adjudicacion',
    'compiledRelease/awards/0/statusDetails': 'detalles_estado',
    'compiledRelease/awards/0/date': 'fecha_adjudicacion_original',
    'compiledRelease/awards/0/value/amount': 'monto_adjudicado',
    'compiledRelease/awards/0/value/currency': 'moneda',
    'compiledRelease/awards/0/invitationID': 'id_invitacion',
    'compiledRelease/awards/0/buyer/id': 'id_institucion_compradora',
    'compiledRelease/awards/0/buyer/name': 'nombre_institucion',
    'compiledRelease/awards/0/relatedBids': 'ofertas_relacionadas_ids',
    'compiledRelease/awards/0/suppliers/0/id': 'ruc_proveedor',
    'compiledRelease/awards/0/suppliers/0/name': 'nombre_proveedor',
    'year': 'anho_adjudicacion'
}

#Renombramos las columnas manteniendo todo el dataset intacto
df_completo_renombrado = df_completo.rename(columns=columnas_nuevas)

# manejamos los casillas vacías
df_completo_renombrado['nombre_institucion'] = df_completo_renombrado['nombre_institucion'].fillna('No especificada')
df_completo_renombrado['anho_adjudicacion'] = df_completo_renombrado['anho_adjudicacion'].fillna(0).astype('int64')

```
